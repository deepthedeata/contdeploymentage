```python
batch_size = 50
img_height = 224
img_width = 224
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory('C:/Users/pr&d02/Desktop/paddy/train', seed = 30, 
                                                              image_size = (img_height, img_width),
                                                              batch_size = batch_size)
```

```python
test_ds = tf.keras.preprocessing.image_dataset_from_directory('C:/Users/pr&d02/Desktop/paddy/test', seed = 30, 
                                                              image_size = (img_height, img_width),
                                                              batch_size = batch_size)
```

```python
validation_ds = tf.keras.preprocessing.image_dataset_from_directory('C:/Users/pr&d02/Desktop/paddy/validation', seed = 30, 
                                                              image_size = (img_height, img_width),
                                                              batch_size = batch_size)
```

```python
for image_batch, labels_batch in train_ds:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
```

```python
classes = train_ds.class_names
print(classes)
```

```python
plt.figure(figsize = (10,10))
for images, labels in train_ds.take(1):
    for i in range(16):
        ax = plt.subplot(4,4,i+1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(classes[labels[i]])
        plt.axis("off")
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().prefetch(buffer_size = AUTOTUNE)
validation_ds = validation_ds.cache().prefetch(buffer_size = AUTOTUNE)
```

```python
feature_extractor = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
```

```python
feature_extractor_layer= hub.KerasLayer(feature_extractor, input_shape=(img_height,img_width,3))
```

```python
feature_extractor_layer.trainable= False  
```

```python
normalization_layer = tf.keras.layers.experimental.preprocessing.Rescaling(1./255)
```

```python
tf.random.set_seed(50)
```

```python
model = tf.keras.Sequential([
    normalization_layer,
    feature_extractor_layer,
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(5,activation = 'softmax') #5 is the number of classes 
])
```

```python
model.compile(
optimizer = 'adam',
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),
metrics = ['accuracy'])
#tf.keras.losses.SparseCategoricalCrossentropy
```

```python
from tensorflow.keras.optimizers.legacy import Adam
```

```python
history = model.fit(train_ds, epochs = 10, validation_data = validation_ds)
```

```python
model.summary()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train_acc','val_acc'], loc = 'best')
plt.show()
```

```python
result = model.evaluate(test_ds)
```
